## 암호학의 기초, 수학

#### 가분성의 의미에 대해 알아봅시다.

$ 만약\ a 와\ b 가\ 정수이고,\ a\neq0\ 일\ 때, \frac{b}{a} = c 가\ 정수이면,\ a\ 가\ b\ 를\ 나눈다고\ 표현하며,\ 이를\ 아래와\ 같이\ 표현$<br><br>
$$a|b\ (반면에,\ b\ 가\ a\ 로\ 나눠지지\ 않는다면\ a \nmid b\ 로\ 표현함)$$

#### 유클리드 호제법을 알아봅시다.
> 두 자연수 a,b 가 있고, a > b 인 경우, a 와 b 의 최대 공약수 는, a 를 b 로 나눈 나머지와 b 의 최대공약수와 같음 
+ <b>증명</b> 
> $두\ 자연수,\ a,b\ (단,\ a > b)\ 의\ 최대공약수를\ gcd(a,b) = G\ 로\ 표현하면,$<br>$$a = AG,\ b=BG\ (단,\ gcd(A,B) = 1)$$<br><br>$그리고,\ gcd(r,b) = g\ 라고\ 했을\ 때,$<br> $$\ r = Rg,\ b = B^{'}g\ (마찬가지로,\ gcd(R, B^{'}) = 1)$$<br><br>$여기서,\ a = bq +r\ (단,\ 1 \leq r < b\ 인\ 자연수)\ 이고,\  동시에\ b = B^{'}g 이므로,$<br>$$\ r = (a-bq) = (a-B^{'}gq) = Rg$$<br><br>$$\therefore,\ a = (R + B^{'}q)g,\ b = B^{'}g$$<br><br>$그런데,\ G=g\ 이기\ 위해선,gcd(R + B^{'}q, B^{'}) = 1\ 이어야\ 함.$<br><br>
$gcd(R + B^{'}q, B^{'}) = M\ 이라고\ 가정$<br>$\therefore, R + B^{'}q = M\alpha,\ B^{'} = M\beta$<br>$\therefore,\ R  = M\alpha - B^{'}q =  M(\alpha - \beta q)$<br> $gcd(R, B^{'}) = 1 이었으므로,\ gcd(R + B^{'}q, B^{'}) = M = 1 $ <br><br> $$\therefore, gcd(a,b) = G = g = gcd(r,b)$$ 
<br><br>
+ 시간 복잡도
> 두 자연수 (n,m) (n > m) 의 최대공약수를 구하기 위한 시간복잡도는<br><br> $$O(\log min(n,m))$$
> + 증명
>> $gcd (n,m) = gcd (r, m)\ (r = n\ 을\ m\ 으로\ 나눈\ 나머지\ r)$<br><br>
>> $n > 2r$<br>
>> $n > 2(n-mq)$<br>
>> $ 2mq > n$<br><br>
>> $여기서,\ 2mq > n\ 는\ 항상\ 성립.$ $$\ \because,\ 나머지\ r\ 에\ 대해,\ 0\leq r< m\ 이기\ 때문에,\ a*mq < n\ 성립\ 불가\ (여기서,\ a\ 는\ 자연수)$$
>> <br>$$ \therefore \  n > 2r\ 이므로,\ \frac{n}{2}>r.\ 유크리드\ 호제법은,\ 최대공약수를\ 구해야하는\ 두\ 수\ 중\ 어느\ 한\ 수가\ 1이\ 될때까지\ 진행.$$ <br>$$\therefore,\ 알고리즘은\ 최악의\ 경우,\ 2 * log\ min(n,m)\ 번\ 진행될\ 것\ (2를\ 곱해준\ 이유는,\ 알고리즘이\ 매회\ 반복될\ 때마다,\ 최댓값(=피제수)이\ 계속\ 바뀌기\ 때문,).$$ 

+ 두 자연수, $a = 66528, b = 52920$ 가 있다. a, b 의 최대공약수를 유클리드 호제법을 사용하여 구하시오. 

```python
def gcd(a,b):
    while b != 0:
        a,b = b, a%b
    return a
```

```python
a=66528
b=52920

gcd(a,b)
```

#### 확장 유클리드 알고리즘에 대해 알아봅시다.
$$a*u + b*v = G = gcd(a,b)\ (단,\ a,b\ 는\ a >b\ 인\ 자연수이고,\ u,v\ 는\ 정수)$$
+ 단계 1.

  > $항상,\ x > d\ 이므로,\ x = dq + r\ 로\ 나타낼\ 수\ 있음.\ (여기서,\ q\ 는\ 몫,\ r\ 은\ 나머지로,\ 0 \leq r < d)$<br><br>
  > $$\therefore,\ x\ 가,\ d\ 의\ 배수가\ 아닌\ 경우가\ 존재하면,\ 0 < r <  d\ 인\ 자연수\ r\ 이\ 반드시\ 존재할\ 것$$<br><br>
  > $r= x-dq$<br>
  > $\;\;= a*u + b*v - q*(a*u_{0} + b*v_{0})$<br>
  > $\;\;= a*u_{k} + b*v_{k}$<br><br>

  > $동시에, 0 < r < d .\ 그런데,\ d\ 가\ S\ 집합\ 내\ 최소의\ 자연수이므로,\ d\ 보다\ 작은\ 자연수\ r\ 은\ 존재할\ 수\ 없음.$<br>

  > $$\therefore\ 집합\ S\ 내에서,\ d\ 보다\ 큰\ 임의의\ 모든\ 자연수들\ x\ 는\ d\ 의\ 배수임.$$
  + $G=gcd(a,b)$ 는 $a*u + b*v$ 로 나타낼 수 있는 최소의 자연수이다
    + 증명  
    > $집합\ S\ 내\ 최소\ 자연수\ 원소를\ d=a*u_{0}+b*v_{0}\ 라고\ 하자.$<br>
    > $또한\ gcd(a,b) = G\ 이므로,\ a=A*G,\ b=B*G\ 라고\ 할\ 수\ 있다.$<br><br>
    > $$\therefore,\ d=a*u_{0}+b*v_{0}$$ $\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;= A*G*u_{0} + B*G*v_{0}$<br>$\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;=G*(A*u_{0} + B*v_{0})\ 라고\ 할수있다.$<br><br> $\ 이는,\ 집합\ S\ 내\ 최소\ 자연수\ d\ 가\ G\ 의\ 배수임을\ 의미한다$<br><br>
    > $동시에,\ a = a*1 + b*0\ 으로\ 나타낼\ 수\ 있고,\ b\ 또한\ 비슷하게\ b = a*0 + b*1\ 로\ 나타낼\ 수\ 있으므로\ 모두\ d\ 의\ 배수들$<br>$\ \therefore\ d\ 는\ a,b\ 의\ 공약수이다.$<br><br>
    > $$종합적으로,\ d\ 는\ G\ 의\ 배수이면서,\ a,b의\ 공약수이다.\ 이러한\ 조건들을\ 만족시키는\ d\ 는\ 명백히\ a\ 와\ b\ 의\ 최대공약수이다$$.$$\ \therefore,\ d=G$$
+ 단계 2.
  > $a = bq_{0} + r_{0}$<br>
  > $b = r_{0}q_{1} + r_{1}$<br>
  > $ r_{0} = r_{1}q_{2} + r_{2}$ <br>
  > $ ... $<br>
  > $r_{i-1} = r_{i}q_{i+1} + r_{i+1}$ <br>
  > $r_{i} = r_{i+1}q_{i+2} + r_{i+2} = d\ 라\ 가정$<br>
  > $\therefore\ r_{i+1} = r_{i+2}q_{i+3}+ r_{i+3} = 0$<br><br>
  > $여기서,\ 점화식\ 유도를\ 위해,\ a = r_{0},\ b =r_{1}\ 이라고\ 할\ 경우,$<br>
  > $a = r_{0} = r_{1}q_{0} + r_{2}$<br>
  > $b = r_{1} = r_{2}q_{1} + r_{3}$<br>
  > $\;\;\;\;\;\;r_{2} = r_{3}q_{2} + r_{4}$<br>
  > $ ... $<br>
  > $\;\;\;\;\;\;r_{i-1} = r_{i}q_{i-1} + r_{i+1}$<br>
  > $\;\;\;\;\;\;r_{i} = r_{i+1}q_{i} + r_{i+2} = d\ 라\ 가정$<br>
  > $\;\;\;\;\;\;r_{i+1} = r_{i+2}q_{i+1}+ r_{i+3} = 0$<br><br>
  > $\therefore\ r_{i+1} = r_{i-1}-r_{i}q_{i-1}$<br><br>
  > $동시에,r_{0} = a = a*1 + b*0,\ r_{1} = b = a*0 + b*1$<br>
  > $\therefore\ r_{i} = as_{i} + bt_{i}\ 로\ 나타낼\ 수\ 있음$<br><br>
  > $r_{i} = as_{i} + bt_{i}\ 의\ 식을\ r_{i+1} = r_{i-1}-r_{i}q_{i-1}\ 에\ 대입하면,$<br>
  > $r_{i+1} = as_{i+1} + bt_{i+1}$<br>
  > $\;\;\;\;\;\;= (as_{i-1} + bt_{i-1}) - (as_{i} + bt_{i})q_{i-1}$<br>
  > $\;\;\;\;\;\;= a(s_{i-1} - s_{i}q_{i-1}) + b(t_{i-1} - t_{i}q_{i-1})$<br><br>
  > $최종적으로,\ r_{i+1} = as_{i+1} + bt_{i+1} = a(s_{i-1} - s_{i}q_{i-1}) + b(t_{i-1} - t_{i}q_{i-1})\ 인\ 점화식을\ 얻을\ 수\ 있고,\ 초항들은\ 아래와\ 같음$<br><br>
  > $s_{0} = 1, t_{0} = 0, s_{1} = 0, t_{1} = 1$<br>
  > $r_{0} = a, r_{1} = b$<br><br>
  > $\therefore,\ 우리는\ r_{i+1}=0\ 일때,\ 그\ 직전\ 항\ r_{i}\ 를\ 찾으면\ 됨 $
+ [추가] 최악의 시간 복잡도
  + 확장 유클리드 알고리즘은, 기존 유클리드 알고리즘에 기반하여,<br><br> $$gcd(a,b) = gcd (r_{0}, b) =gcd(r_{0},r_{1}) = gcd(r_{2},r_{1}) = ... gcd(r_{i},r_{i+1}) = d $$<br><br> 의 $r_{i+1} = a*u + b*v = 0$ 일 때의 정수 $(u, v)$ 쌍을 단계 2의 점화식을 이용하여 구하는 것임.<br><br> 
  + $\therefore$ 최악의 시간복잡도는, 기존의 유클리드 알고리즘과 동일한 $O(log\ min(a,b))$ 임. 

+ 두개의 소수 $p=26513,\ q=32321$ 에 대하여, $p*u\ +\ q*v=gcd(p,q)$ 를 만족시키는 최소 정수 쌍 $(u,v)$ 를 제출하시오.
  + $p$ 와 $q$ 는 소수로, 서로소 관계에 있음. $$\therefore,\ p*u\ +\ q*v=26513*u\ +\ 32321*v= 1 $$

```python
def egcd(a,b): 
    s_prev, s_now = 1,0
    t_prev, t_now = 0,1 
    
    r_prev, r_now, r_next = a, b, 1
    
    while r_next : 
        s_next = s_prev - s_now*(r_prev//r_now)
        t_next = t_prev - t_now*(r_prev//r_now)
        r_next = a*s_next + b*t_next
        
        s_prev, s_now = s_now, s_next
        t_prev, t_now = t_now, t_next
        r_prev, r_now = r_now, r_next
            
    return s_prev, t_prev
```

```python
p = 26513
q = 32321

egcd(p,q)
```

#### 모듈러 연산에 대해 알아봅시다.

+ 모듈러 연산<br>: 정수 $a$ 를 양의 정수 $b$ 로 나눈 나머지가 r ($ 0 \leq r < b $) 일 때, 나머지 r 을 다음과 같이 표현<br><br> $$ a\mod b = r$$ 

+ 모듈러 합동 <br>: 두 정수 $a,\ b$ 가 있을 때, $a-b$ 를 $m$ 으로 나눈 나머지가 0 인 경우, 즉, $a-b$ 가 $m$ 으로 나눠 떨어지는 경우, 정수 $a$ 와 $b$ 를 모듈러 $m$ 합동 ($a\ is\ congruent\ to\ b\ modular\ m $)이라고 하고, 이를 아래와 같은 합동 기호($\equiv$)를 사용하여 표현<br><br>$$a \equiv b (\mod m)$$<br><br>
   + [정리 1] 두 정수 a,b 에 대해, $a \equiv b\ (\mod\ m)$ 은, $a \mod m = b \mod m$ 와 필요충분 조건 관계에 있음.
      + 증명
      > $a \mod m = b \mod m = r\ 이\ 성립한다고\ 가정.$<br>
      > $\therefore a = mq_{0} + r,$<br>$\;\;\; b = mq_{1} + r$<br><br>
      > $\therefore a - b = (mq_{0} + r) - (mq_{1} + r)$<br> 
      > $\therefore \;\;\;\;\;\;\;\;\; = m(q_{0} - q_{1})$<br><br>
      > $$\therefore a \mod m = b \mod m\ 인\ 경우,\ a \equiv b\ (\mod\ m)\ 항상\ 성립$$<br><br>
      > $a = mq_{0} + r_{0},\; b =mq_{1} + r_{1}\ 이라고\ 가정$<br>
      > $\therefore\ a - b = m(q_{0} - q_{1}) + (r_{0} - r_{1})$<br><br>
      > $여기서, a \equiv b\ (\mod\ m) 이므로,\ a-b = m(q_{0} - q_{1}) + (r_{0} - r_{1}) = mq\;\; (q_{0}, q_{1}, q\ 는\ 모두\ 정수)$<br>
      > $그런데, 0 \leq r_{0}<m,\; 0 \leq r_{1}<m 이므로, |r_{0}-r_{1}| < m$<br><br>
      > $$\therefore, a \equiv b\ (\mod\ m)\ 인\ 경우, a \mod m = b \mod m\ 항상\ 성립 $$
   + [정리 2] 양의 정수 m 이 주어졌을 때, $a \equiv b (\mod m)$, $c \equiv d (\mod m)$ 이 성립하면, $a+c \equiv b+d (\mod m)$  과 $ac \equiv bd (\mod m)\ 이\ 항상\ 성립$ 
      + 증명
      > $먼저,\ 두\ 정수\ a\ 와\ b\ 가\ 모듈러\ 합동인\ 경우,\ a-b = mq\ 인\ 정수\ q\ 가\ 존재함을\ 의미$<br>
      > $이는,\ 정수\ a\ 를,\ b+mq\ (q\ 는\ 정수)\ 의\ 꼴로\ 나타낼\ 수\ 있음을\ 의미$<br><br>
      > $\therefore\ b = a+ mq,\; d = c + mp\ (단,\ q,p\ 모두\ 정수)$<br><br>
      > $그리고,\ b+d= a+ mq + c + mp = (a+c) + m(q+p)$<br><br>
      > $$\therefore\ a+c \equiv b+d\ (\mod m) $$<br><br>
      > $그리고,\ b \times d = ac + m (mpq + qc + pa)$<br><br>
      > $$\therefore\ ac \equiv bd\ (\mod m) $$
   + [정리 3] $m$ 이 양의 정수, $a,b$ 가 정수이면 다음이 성립.
     $$(a+b) \mod m = ((a \mod m) + (b \mod m)) \mod m $$$$ab \mod m = ((a \mod m)(b \mod m)) \mod m $$<br><br>
      + 증명
      > $[정리 1]\ 에\ 따라, a \equiv (a \mod m) (\mod m),\;\; b \equiv (b \mod m) (\mod m)$<br><br>
      > $[정리 2]\ 에\ 따라, a+b \equiv ((a \mod m) + (b \mod m)) (\mod m)$
      > $\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\; ab \equiv ((a \mod m)(b \mod m)) (\mod m) $

+ 아래의 두 합동식을 만족시키는 가장 작은 양의 정수 (x,y) 쌍을 제출하시오<br><br>
$$11 \equiv x (\mod 6),\;\; 8146798528947 \equiv y (\mod 17)$$<br><br>
  + x,y 는 위의 합동식을 만족시키는 가장 작은 양의 정수임에 주목하라 ([정리 1] 사용할 것)

#### 페르마의 소정리에 대해 알아봅시다.
+ 페르마의 소정리
  + 증명 (간략화)
  > 정수 a, 소수 p 가 있다고 가정. 그리고, 정수 a 는 소수 p 로 나눠지지 않음.
  > $$\therefore\ a = pq + r\ (q\ 는\ 정수,\ 0 < r< p\ 인\ 정수)$$<br><br>
  > + 단계1. $정수 1 \times a, 2 \times a, 3 \times a, ..., (p-1)a\ 중\ 어떤\ 두\ 수도\ 모듈로\ p\ 와\ 합동관계에\ 있지\ 않음$
  >> 증명 <br><br>
  >> $서로\ 다른\ 두수\ ka = k(pq +r),\ la = l(pq+r)\ 가정 (단, 1 \leq k\neq l \leq p-1\ 인\ 자연수)$<br><br>
  >> $모듈러\ 합동의\ [정리 3]\ 에\ 의해,\ ka \mod p,\ la \mod p\ 을\ 다음과\ 같이\ 정리\ 가능$
  >> $$ka \mod p = r(k \mod p)\mod p$$$$la \mod p = r(l \mod p)\mod p$$<br><br>
  >> $l, k 각각이 1 \leq k,l \leq p-1\ 인\ 자연수 였으므로,  (l \mod p) \neq (k \mod p)$<br><br>
  >> $$\therefore\ ka,\ la\ 은\ 모듈러\ p\ 에\ 대해\ 합동관계에\ 있지\ 않음$$
  > + 단계2. $단계\ 1 의\ 결과를\ 이용하여,\ 1,\ 2,\ ...,\ p-1\ 의\ 곱이\ a,\ 2a,\ ..., (p-1)a\ 의\ 곱에\ 모듈러\ p로\ 합동임을\ 유도하라.\ 그리고,\ 이를\ 이용하여\ 다음을\ 증명하라.$<br>
     $$(p-1)! \equiv a^{p-1}(p-1)! (\mod p)$$ 
  >> 증명 <br><br>
  >> $ 1 \leq k \leq p\ 인\ 자연수\ k\ 에\ 대해,\ 1 \leq (ka \mod p) \leq p-1$
  >> $ 동시에,\ 단계1\ 에서,\ a,\ 2a,\ ...\ (p-1)a\ 모두\ 모듈러\ p\ 에\ 대해\ 모두\ 다른\ 값을\ 가진다고\ 했음.$<br><br>
  >> $$\therefore, (a \mod p) \times (2a \mod p) \times ... \times ((p-1)a \mod p) = a^{p-1}(p-1)! \mod p = (p-1)! \mod p $$<br><br>
  >> $$\therefore, (p-1)! \equiv a^{p-1}(p-1)! (\mod p)$$
  > + 단계3. $단계\ 2 의\ 결과를\ 이용하여,\ p \nmid a\ 인\ 경우,\ a^{p-1} \equiv 1 (\mod p)\ 임을\ 유도하라$<br>
  >> 증명 <br><br>
  >> $(p-1)! = 1 \times 2 \times 3 ... \times (p-1)\ 로,\ 어떠한\ 두\ 개\ 이상의\ 양의\ 정수들의\ 곱,\ 혹은,\ 그\ 어떠한\ 양의\ 정수도\ 소수\ p\ 의\ 약수가\ 될\ 수\ 없음$<br>
  >> $$\therefore\ gcd((p-1)!, p) = 1$$<br><br>
  >> + 다음의 보조 정리들 필요
  >>   + [보조 정리1]양의 정수 $a,b,c\ 가\ gcd(a,b) = 1$ 이고,$\ a | bc\ (bc\ 가\ a\ 로\ 나눠짐.\ 나머지 = 0\ 을\ 의미)$ 이면,$\ a|c$ 이다.
  >>> $gcd(a,b) = au+bv = 1\ 의\ 양변에\ 양의\ 정수\ c\ 를\ 곱함$

  >>   + [보조 정리2]m 이 양의 정수이고, a,b,c 를 정수라고 하자. $ac \equiv bc (\mod m)\ 이고,\ gcd(c,m) = 1\ 면,\ a \equiv b (\mod m)\ 성립$

  >>> $$\therefore m |(a - b)\ 이므로,\ 모듈러\ 합동의\ 정의에\ 따라,\ a \equiv b (\mod m)$$

#### 모듈러 역수에 대해 알아봅시다.
+ <b>모듈러 역수 (Modular Inverse)</b><br> : 서로소 관계에 있는 정수 a 와, 양의 정수 p 에 대해,  $a$ 의 양의 정수 $p$ 에 대한 모듈러 역수는 아래와 같이 표현 가능<br><br>
  $$1 \equiv a \times a^{-1} (\mod p) $$<br><br>
  + 모듈러 $a (\mod p)$ 에서, $a$ 의 양의 정수 $p$ 에 대한 모듈러 역원이 존재하기 위한 필요충분 조건은 $gcd(a,p)=1$ 임.
    + 증명 (간략화)
    > $a\ 의\ 양의\ 정수\ p\ 에\ 대한\ 모듈러\ 역원을\ a^{-1}\ 이라고\ 하자.$<br>
    > $그렇다면,\ 모듈러\ 역원의\ 정의에\ 따라\ , (a \times a^{-1}) \mod p = 1$<br><br>
    > $만약,\ gcd(a,p) \neq 1\ 이라면,\ a = kA,\ p = kP\ 로\ 나타낼\ 수\ 있을\ 것이다.\ (여기서,\ 정수\ k\ 는,\ 정수\ a\ 와\ p\ 의\ 공통\ 인수이다.)$ <br><br>
    > $\therefore,\ (a \mod p) = (kA \mod p)= kA -pq_{0} = kA -kPq_{0} = k(A-Pq_{0}) = r (단,\ 0 \leq r \leq p-1\ 인\ 자연수, q\ 는\ 정수)$<br><br>
    > $\therefore,\ (a \times a^{-1}) \mod p =(a \mod p) (a^{-1} \mod p) \mod p$<br>
    > $\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;= r (a^{-1} \mod p) \mod p$<br> 
    > $\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;= rI -pq_{1} \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\; (편의상,\ (a^{-1} \mod p) = I\ 로\ 표현)$<br> 
    > $\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;=k(A−Pq_{0})I-kPq_{1}$<br>
    > $\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;=k((A−Pq_{0})I - Pq_{1})$<br><br>
    > $여기서,\ k는\ 1\ 또는\ 0\ 이\ 아닌\ 양의\ 정수이고, A, P, q_{0}, q_{1} 과 I\ 는\ 모두\ 정수이다.$<br>
    > $\therefore,\ 반드시\ k((A−Pq_{0})I−Pq_{1}) \neq 1 이게 됨.$<br><br>
    > $$\therefore gcd(a,p) = 1\ 인\ 경우만,\ 모듈러\ 역원이\ 존재함.$$

+ 아래 두가지 방법을 이용하여, 아래의 모듈러 합동식에, 3 의 소수 13 에 대한 모듈러 역원을 구하라.<br><br> $$3 \times d \equiv 1 (\mod 13)$$<br><br>
   + 페르마 소정리
     > $gcd (3,13) = 1\ 이므로,\ 3^{12} \mod 12 = (3 \times 3^{11}) \mod 13$<br>
     > $\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;=3 (3^{11} \mod 13) \mod 13$<br>
     > $\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;= 1$<br><br>
     > $$\therefore,\ 우리는\ 3^{11} \mod 13\ 값을\ 구하면\ 됨$$
   + 확장 유클리드 호제법
     > $gcd(3,13) = 3u + 13v =1 $<br><br>
     > $(3u + 13v) \mod 13 = (3u \mod 13 + 13v \mod 13) \mod 13$<br>
     > $\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\; = (3u \mod 13) \mod 13$<br>
     > $\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\; = ((3 \mod 13) (u \mod 13)) \mod 13$<br>
     > $\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\; = 1$<br><br>
     > $$\therefore,\ 우리는\ 확장유클리드\ 알고리즘을\ 이용하여,\ u\ 를\ 구한\ 뒤,\ u \mod 13 값을\ 구하면\ 됨$$

```python
## 페르마 소정리 이용
## gcd (3,13) = 1 이므로, 3^12 mod 12 = 3*3^11 mod 13  = 1
## 3*3^11 mod 13 = 3 (3^11 mod 13) mod 13 =  1

3**11 % 13
```

```python
## 확장 유클리드 호제법 사용
## 3u + 13v = 1
## (3u + 13v) mod 13 = 1
## (3u mod 13 + 13v mod 13) mod 13
##   = (3u mod 13) mod 13
##   = (3u mod 13) = ((3 mod 13) (u mod 13)) mod 13 = 1

## 위의 식을 만족시키는 13 보다 작은 음이 아닌 정수를 선택 (u 를 구한 다음, u mod 13)

egcd(3,13)[0] % 13
```
