## 암호를 위한 심화 수학

#### 제곱잉여에 대해 알아봅시다.
+ 제곱 잉여 (Quadratic Residue)<br>: 정수 $a$, 양의 정수 $p$ 에 대해, $x^2 \equiv a (\mod p)$ 가 성립하는 정수 $x$ 가 있을 경우, $a$ 를 $p$ 의 <b>제곱 잉여</b>라고 함
  > 이와는 반대로, $x^2 \equiv a (\mod p)$ 가 성립하는 정수 $x$ 가 없는 경우, $a$ 를 $p$의 <b>제곱 비잉여</b>라고 함
+ 2 보다 큰 양의 정수 p 


  > $(p-b)^{2} = p^2-2pb+b^2 \equiv b^2 (\mod p) $<br><br>
  > $\therefore\; 1^2, 2^2, ... , (p-2)^{2}, (p-1)^{2}$
  > $(\alpha)$
+ $2$ 보다 큰 소수 $p$ 와, $1 \leq a \leq p-1$ 인 모든 정수 $a$ 에 대하여, $p$ 의 제곱 잉여, 제곱 비잉여는 각각 약 $\frac{(p-1)}{2}$ 개임.
  > $1 \leq a \leq p-1\ 의\ 모든\ 정수들에\ 대한\ 제곱수들을\ 아래와\ 같이\ 가정하자.$<br>
  > $$1^2, 2^2, ... , (p-2)^{2}, (p-1)^{2}$$<br><br>
  > $위의\ 제곱수들은, 1 \leq b \leq (p-1)\ 인\ 정수\ b\ 를\ 이용하여,\ p-b 의\ 꼴로\ 나타낼\ 수\ 있음$<br><br>
  > $그런데,\ (p-b)\ 에\ 대한\ 모듈러\ p\ 는\ 아래와\ 같이\ 유도\ 가능$<br>
  > $$(p-b)^{2} = p^2-2pb+b^2 \equiv b^2 (\mod p)$$<br><br>
  > $\therefore\ 위의\ 제곱수들에서\ 같은\ 모듈러\ p\ 값을\ 가지는\ 제곱수들끼리를\ 지워보면,\ 아래의\ 제곱수들만\ 남는다는\ 것을\ 알\ 수\ 있음.$<br>
  > $$1^2, 2^2, ..., (\frac{p-1}{2})^{2} $$<br><br>
  > + 이제, $p-1$ 개의 제곱수들($1^2, 2^2, ..., (\frac{p-1}{2})^{2}$) 중, 그 어떠한 두수도 모듈러 $p$ 에 대한 합동이 아님을 증명<br><br>
  >> $서로\ 다른\ 두\ 제곱수\ \alpha^2,\ \beta^2\ (1 \leq \alpha, \beta \leq p-1)\ 간에\ 아래와\ 같은\ 합동관계가\ 존재한다고\ 가정.$<br>
  >> $$\alpha^2 \equiv \beta^2 (\mod p))$$<br><br>
  >> $\therefore\ 아래와\ 같은\ 가분성\ 성립$<br>
  >> $\;\;\;\;p | (\alpha^2 - \beta^2)$<br><br>
  >> $p\ 는\ 소수이므로,\ p|(\alpha^2 - \beta^2)\ 이\ 성립하기\ 위해서는\ 반드시\ p|(\alpha- \beta)\ 혹은,\ p|(\alpha + \beta)\ 중\ 하나가\ 성립$<br>
  >> $그런데,\ \alpha-\beta,\ \alpha+\beta\ 모두\ 소수\ p\ 보다\ 작음.$<br>
  >> $\therefore p\nmid(\alpha- \beta)\; 면서,\; p\nmid(\alpha+\beta)$<br><br>
  >> $이와\ 같은\ 모순은,\ 어떤\ 서로\ 다른\ 두\ 제곱수\ \alpha^2,\ \beta^2\ 간에\ 모듈러\ p에\ 대한\ 합동관계가\ 존재한다는\ 가정이\ 틀렸기\ 때문$<br><br>
  > $\therefore, $
